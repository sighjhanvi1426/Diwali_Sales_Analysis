# Exploratory Data Analysis

### Analysis based on Gender

```python
Gender = sns.countplot(x = 'Gender' , data = Data , color = 'Red')

# Adding bar labels
for bars in Gender.containers:
    Gender.bar_label(bars, padding=3, fmt='%d', fontsize=10 )

# Customizing the plot
plt.title('Gender Distribution', fontsize=16, fontweight='bold', color='Blue')
plt.xlabel('Gender', fontsize=12, fontweight='bold' , color = 'Black')
plt.ylabel('Count', fontsize=12, fontweight='bold' , color = 'Black')

# Adjusting the layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Analysie the data based on the gender where we get to know how much amount spend by the male as well as female.

Comparsion = Data.groupby(['Gender'] ,as_index = False ) ['Amount'].sum().sort_values(by = 'Amount' , ascending = False )
Comparsion
```

```python
sns.barplot(x = 'Gender'  , y = 'Amount' , data = Comparsion , color = 'yellow')

# Adding bar labels
for bars in plt.gca().containers:
    plt.gca().bar_label(bars, padding=3, fmt='%d', fontsize=10)

# Customizing the plot
plt.title('Gender Distribution By Amount', fontsize=16, fontweight='bold', color='Blue')
plt.xlabel('Gender', fontsize=12, fontweight='bold' , color = 'Black')
plt.ylabel('Amount', fontsize=12, fontweight='bold' , color = 'Black')

# Adjusting the layout
plt.tight_layout()

# Display the plot
plt.show()
```

### Analysis based on Age group

```python
# Ascending Order of Age Group
Age_order = sorted(Data['Age Group'].unique())

# Creating the Count Plot
Age = sns.countplot(x = 'Age Group' , data = Data , hue = 'Gender' , order = Age_order)

# Adding the bar lables 
for bars in Age.containers:
    Age.bar_label(bars)

# Customizing Plot Titles and Labels
plt.title('Age Group Distrubtion by Gender' , fontsize = 15 , fontweight = 'bold')
plt.xlabel('Age Group' , fontsize = 11 , fontweight = 'bold')
plt.ylabel('Count' , fontsize = 11 , fontweight ='bold')
plt.legend(title = 'Gender')

# Adjusts the layout to prevent overlap of plot elements, ensuring everything fits within the figure boundaries.
plt.tight_layout()

#Display the plot
plt.show()
```

```python
#Analysie the data based on the age group where we get to know how much amount spend by the that age

Age_comparsion = Data.groupby(['Age Group'] , as_index = False) ['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
Age_comparsion
```

```python
sns.barplot(x = "Age Group" , y = "Amount" , data = Age_comparsion , hue="Age Group", palette='Set2')

sns.set(rc={'figure.figsize':(15,6)})
plt.title ('Age Group by Amount' , fontsize = 15 , fontweight = 'bold')
plt.xlabel ('Age Group' , fontsize = 11 , fontweight = 'bold' , color = 'black')
plt.ylabel ('Amount' , fontsize = 11 , fontweight = 'bold' , color = 'black')

plt.tight_layout()

plt.show()
```

### Analysis based on States

```python
States = Data.groupby(['State'], as_index = False) ['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
States
```

```python
# All States
sns.set(rc={'figure.figsize':(22,6)})
sns.barplot(data = States , x ='State' , y = 'Orders' , hue="State", palette='Set2')




# Customize the plot
plt.title('Orders by State', fontsize=20, fontweight='bold')
plt.xlabel('States', fontsize=15, fontweight='bold')
plt.ylabel('Number of Orders', fontsize=15, fontweight='bold')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
